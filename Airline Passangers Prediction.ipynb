# Spliting into X AND Y

```python
x = df.drop('satisfaction_satisfied',axis=1)
y = df['satisfaction_satisfied']
```

```python
x
```

```python
y
```

# Train_test_split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

# Feature Scalling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_train = scaler.transform(x_train)
```

# Creating Artifical Neural Network

```python
#!pip install keras
```

```python
# Importing Libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
x_train.shape
```

```python
# Initializing the ANN
ann_clf = Sequential()
```

```python
#Adding input layer and first hidden layer
ann_clf.add(Dense(activation = 'relu', input_dim = 23, units = 6, kernel_initializer='uniform'))
```

```python
# Adding Second Hidden Layer
ann_clf.add(Dense(activation ='relu', units = 6, kernel_initializer = 'uniform'))
```

```python
# Adding Output Layer
ann_clf.add(Dense(activation = 'sigmoid',units = 1, kernel_initializer = 'uniform'))
```

```python
# Model Compilation
ann_clf.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# Model Summary

```python
ann_clf.summary()
```

# Model Training

```python
ann_clf.fit(x_train, y_train, validation_split = 0.33, batch_size = 32,epochs = 50)
```

```python
# Predicting the test result
y_pred = ann_clf.predict(x_test)
y_pred
y_pred = (y_pred>0.5)
```

# Accuracy and Confusion Matrix

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
acc = accuracy_score(y_test,y_pred)
```

```python
acc
```

# Confusion Matrix

```python
cm = confusion_matrix(y_test,y_pred)
cm
```
